### SQL Ch. 4 - Cast (and other column value manipulation)

```python
from IPython.display import Image
```

```python
# what is the inventory value of each product

SELECT ProductName, RetailPrice * QuantityonHand as inventoryValue
from salesordersexample.Products
```

```python
# days elapsed

SELECT OrderNumber, OrderDate, ShipDate, CAST(ShipDate - OrderDate AS Integer) as DaysElapsed
FROM salesordersexample.Orders
```

```python
Image(filename='capture.png') 
```

```python
# how long will each engagement run?
SELECT EngagementNumber, CAST(CAST(EndDate - StartDate AS Integer) + 1 AS Character) || ' day(s) ' AS DueToRun
FROM entertainmentagencyexample.Engagements
```

```python
# what if we adjusted each product price by reducing it 5%
SELECT product_vendors.productnumber,
    product_vendors.wholesaleprice,
    product_vendors.wholesaleprice - product_vendors.wholesaleprice * 0.05 AS newprice
   FROM salesordersexample.product_vendors;
```

```python
# show a list of orders made by each customer in desc date order
SELECT customerid, orderdate
FROM salesordersexample.orders
ORDER BY customerid, orderdate
```

```python
#  compile a complete list of vendor names and addresses in vendor name order
select vendname, vendstreetaddress, vendcity, vendstate, vendzipcode
from salesordersexample.vendors
order by vendname 
```

```python
# give me the names of all our customers by city
SELECT custfirstname || ' ' || custlastname AS CustName, custcity
FROM entertainmentagencyexample.customers
ORDER BY custcity
```

```python
# show the date of each agent's first 6-month performance review
SELECT agtfirstname || ' ' || agtlastname AS agtname, datehired, datehired + INTERVAL '6 Months' AS perfreview
FROM entertainmentagencyexample.agents
```
