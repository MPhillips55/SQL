Filter data using the 'WHERE' clause

```python
SELECT "columns"
FROM "table"
WHERE "search condition"
```

```python
# Example
SELECT CustLastName
FROM Customers
WHERE CustLastName = 'Smith'
```

5 types of search conditions generally used:
- Comparison: = (equal to), <> (not equal to), <, <=, >, >=
- Between: does the value fall within a range? specify the range using two value expressions separated by AND keywork, includes the start and end points in the result
- In (Membership): test whether the value matches an item in a given list 
- Like (pattern match): test whether a character string matches a specified pattern (% represents any number of extra characters, _ represents 1 character)
- Is Null: is the value null?

```python
# Range Exmaple
SELECT firstname, lastname
FROM Staff
WHERE DateHired
BETWEEN '1986-07-01' AND '1986-07-11'
```

```python
# Range for character strings, starts with "b"
SELECT studlastname, studfirstname
FROM students
WHERE studlastname BETWEEN 'B' AND 'Bz'
```

```python
# Set example
SELECT TourneyLocation
FROM Tournaments
WHERE TourneyDate
    IN ('2017-9-18', '2017-10-09')
```

```python
# Pattern Match example, custlastname begins with Mar
SELECT custlastname, custfirstname
FROM Customers
WHERE custlastname LIKE 'Mar%'
```

```python
# Pattern Match, escape example
SELECT productname, productcode
FROM Products
WHERE productcode LIKE 'G\_00_' ESCAPE '\'
```

```python
# Is Null example
SELECT custfirstname || ' ' || custlastname AS Customer
FROM Customers
WHERE custcountry IS NULL
```

```python
# Not example, all orders not in october
SELECT orderid, orderdate
FROM Orders
WHERE orderdate NOT BETWEEN '2017-10-01' AND '2017-10-31'
```

    AND/OR allow you to check for multiple conditions

```python
# AND example
SELECT custfirstname, custlastname
FROM Customers
WHERE custcity = 'Seattle' AND custlastname LIKE 'H%'
```

```python
# OR example
SELECT stffirstname, stflastname, stfcity, stfstate
FROM Staff
WHERE stfcity = 'Seattle' OR stfstate = 'OR'
```

```python
# AND and OR combined example
SELECT stffirstname, stflastname, stfareacode, datehired
FROM Staff
WHERE (stfareacode = '425' AND stfphonenumber LIKE '555%')
    OR datehired 
    BETWEEN '2017-10-1' AND '2017-12-31'
```
